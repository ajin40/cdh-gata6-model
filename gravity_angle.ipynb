```python
import numpy as np

def slope_angle(x, y, curvature):
    return np.sin(curvature * (2 * x * np.cos(np.arctan2(y, x))+ 2 * y * np.sin(np.arctan2(y, x))))

def paraboloid(x,y,a):
    return a*(x**2 + y**2)

def paraboloid_slope(x, y, a):
    return [2*a*x, 2*a*y, -1]/np.linalg.norm([2*a*x, 2*a*y, -1])
```

```python
from matplotlib import pyplot as plt

x = np.linspace(-30, 30, num=600)
y = np.linspace(-30, 30, num=600)
X, Y = np.meshgrid(x,y)
z = paraboloid(X, Y, 1/60)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, z, 50)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D contour')
plt.show()

print(z[-1, -1])
print(z[0, -1])
print(z[0, 0])
print(z[-1, 0])
```

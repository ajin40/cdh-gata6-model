```python
# Colab setup ------------------
import os, sys, subprocess
# ------------------------------

import numpy as np
import scipy.integrate

import bokeh.io
import bokeh.layouts
import bokeh.models
import bokeh.plotting

import GATA6_response_model_v2

fox_max = 400
num_entries=11
max_gata6 = 10000

fox_max = 400
num_entries=11
max_gata6 = 10000
gata6_values = np.linspace(0, max_gata6, num_entries)
foxa2_t0 = np.linspace(0, fox_max, num_entries)
foxf1_t0 = np.linspace(0, fox_max, num_entries)
foxa2s = np.zeros((len(gata6_values), len(foxa2_t0), len(foxf1_t0), 10001))
foxf1s = np.zeros((len(gata6_values), len(foxa2_t0), len(foxf1_t0), 10001))
for gata6 in range(len(gata6_values)):
    for foxa2 in range(len(foxa2_t0)):
        for foxf1 in range(len(foxf1_t0)):
            a, b, c = GATA6_response_model_v2.RunDeterministicResponse(GATA6=gata6_values[gata6], T=10000, dt=1., plots=False, plotPromoter=False,
                                                                        inhibited_transcription_rate=[5e-8, 5e-8],
                                                                        activated_transcription_rate = [0.01, 0.01],
                                                                        double_activated_transcription_rate= [0.015, 0.015],
                                                                        triple_activated_transcription_rate = [0.05, 0.025],
                                                                        quad_activated_transcription_rate = [0.1, 0.05],
                                                                        P_act_binding = [0.025/100, 0.05/100],
                                                                        P_inh_binding=[100.0/100, 5.0/100],
                                                                        P1_act_unbinding=[100.0/100, 100.0/100],
                                                                        P1_inh_unbinding=[20.0/100, 1.0/100],
                                                                        P1_act_binding=[0.5/100, 0.7/100],
                                                                        P2_act_unbinding=[100.0/100, 100.0/100],
                                                                        P2_act_binding=[1.0/100, 1.0/100],
                                                                        P3_act_unbinding=[100.0/100, 100.0/100],
                                                                        P3_act_binding=[1.0/100, 1.0/100],
                                                                        P4_act_unbinding=[100.0/100, 100.0/100],
                                                                        FOXA2_prot=foxa2_t0[foxa2],
                                                                        FOXF1_prot=foxf1_t0[foxf1])
            foxa2s[gata6, foxa2, foxf1, :] = c[:,2]
            foxf1s[gata6, foxa2, foxf1, :] = c[:,3]
```

```python
notebook_url="localhost:8888"
bokeh.io.output_notebook()

               
def callback(attr, old, new):
    for i in range(len(foxa2_t0)):
        for j in range(len(foxf1_t0)):
            cd_array[i,j].data['foxa2'] = foxa2s[int(gata6_slider.value), i, j, :]
            cd_array[i,j].data['foxf1'] = foxf1s[int(gata6_slider.value), i, j, :]
            
            
gata6_slider = bokeh.models.Slider(
    title='GATA6 Value', start=0, end=num_entries-1, step=1, value=0, width=150
)


cds = []
for i in range(len(foxa2_t0)):
    for j in range(len(foxf1_t0)):
        cds.append(bokeh.models.ColumnDataSource(dict(foxa2 = foxa2s[gata6_slider.value, i, j, :], foxf1=foxf1s[gata6_slider.value, i, j, :])))
cd_array = np.array(cds).reshape((len(foxa2_t0), len(foxf1_t0)))
# Build s, taking slider values are parameters.

a = bokeh.plotting.figure(
    frame_height=300,
    frame_width=300,
    x_axis_label='FOXA2',
    y_axis_label='FOXF1',
    x_range=[-0.02, fox_max+0.02],
    y_range=[-0.02, fox_max+0.02]
)


for foxa2 in range(len(foxa2_t0)):
    for foxf1 in range(len(foxf1_t0)):
        if foxa2s[gata6_slider.value, foxa2, foxf1, -1] > foxf1s[gata6_slider.value, foxa2, foxf1, -1]:
            a.line(source=cd_array[foxa2,foxf1], x='foxa2', y='foxf1')
        else:
            a.line(source=cd_array[foxa2,foxf1], x='foxa2', y='foxf1')
    
layout = bokeh.layouts.column(
    a, gata6_slider
)

gata6_slider.on_change("value", callback)

def app(doc):
    doc.add_root(layout)
```

```python
interactive_python_plots = True
if interactive_python_plots:
    bokeh.io.show(app, notebook_url=notebook_url)
```
